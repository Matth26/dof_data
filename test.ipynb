```python
import pandas as pd
name = "Amublop_Griotte_Royale"
df1 = pd.read_csv("./import_data/obj_csv/" + name + ".csv",index_col=0)
df1.index = pd.to_datetime(df1.index, format="%d/%m/%Y %H:%M:%S")
df1
```

```python
import matplotlib.pyplot as plt
df1.plot(y=['average_sold_price', 'min_selling_price'], use_index=True, title=name)
plt.gcf().autofmt_xdate()
```
